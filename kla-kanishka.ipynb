```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
img1 = cv2.imread("wafer_image_1.png")
img2=cv2.imread("wafer_image_2.png")
img3=cv2.imread("wafer_image_3.png")
img4=cv2.imread("wafer_image_4.png")
img5=cv2.imread("wafer_image_5.png")
cv2_imshow(img1)
cv2.waitKey(0)

# Use the cvtColor() function to grayscale the image
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2GRAY)
img4 = cv2.cvtColor(img4, cv2.COLOR_BGR2GRAY)
img5 = cv2.cvtColor(img5, cv2.COLOR_BGR2GRAY)
print("\n Grey Scaling\n")
cv2_imshow( img1)
cv2.waitKey(0)
# Window shown waits for any key pressing event
cv2.destroyAllWindows()
print("Image 2")
cv2_imshow(img2)
cv2.waitKey(0)

def mse(imageA, imageB):
	err = np.sum((imageA.astype("float") - imageB.astype("float")) ** 2)
	err /= float(imageA.shape[0] * imageA.shape[1])
	return err

def compare_images(img1, img2, title):

	m = mse(img1, img2)
	s = ssim(img1, img2)
	fig = plt.figure(title)
	plt.suptitle("MSE: %.2f, SSIM: %.2f" % (m, s))

	# first image
	ax = fig.add_subplot(1, 2, 1)
	plt.imshow(img1, cmap = plt.cm.gray)
	plt.axis("off")

	# second image
	ax = fig.add_subplot(1, 2, 2)
	plt.imshow(img2, cmap = plt.cm.gray)
	plt.axis("off")
	plt.show()

compare_images(img1,img2,"image")
```
